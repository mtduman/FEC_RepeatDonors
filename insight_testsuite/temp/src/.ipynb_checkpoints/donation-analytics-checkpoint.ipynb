```python
import sys
import pandas as pd


def aggregate_TXT(txt_fname):
    header_names = ['CMTE_ID','NAME','ZIP_CODE','TRANSACTION_DT','TRANSACTION_AMT','OTHER_ID']
    column_type = {'TRANSACTION_AMT':int}
    
    df = pd.read_csv( txt_fname, delimiter='|', usecols=[0,7,10,13,14,15], dtype=str, names=header_names, converters=column_type)

    df = df[ ~(df['OTHER_ID'].notnull() |
               df['NAME'].isnull() |
               df['CMTE_ID'].isnull() |
               df['TRANSACTION_AMT'].isnull() )]

    df['ZIP_CODE'] = df['ZIP_CODE'].str[:5]
    df['TRANSACTION_DT']  = pd.to_datetime(df['TRANSACTION_DT'] ,format='%m%d%Y' , errors='coerce' )

    df = df[~( (df['ZIP_CODE'].str.len() != 5) |
                df['TRANSACTION_DT'].isnull()    ) ]

    df['ID'] = df.index
    return df


def run(fcont,fperc,fout):
#     if fcont == '':
#         fnames = [ '/Users/ekinezgi/Downloads/itcont_test.txt']
    #     fnames = [  '/Users/ekinezgi/Downloads/indiv14/itcont.txt',
    #                 '/Users/ekinezgi/Downloads/indiv16/itcont.txt',
    #                 '/Users/ekinezgi/Downloads/indiv18/itcont.txt']
    #     fnames = [ '/Users/ekinezgi/Downloads/indiv18/itcont.txt']
    
#     if fperc == '':
#         fperc = '/Users/ekinezgi/Downloads/percentile_test.txt'
#     if fout == '':
#         fout = '/Users/ekinezgi/Downloads/repeat_donors.txt'

    
    actYear = pd.Timestamp("today").year
    actPercentile = int( open(fperc).readline().rstrip() )
    
    
    df = aggregate_TXT(fcont)
    df['count_all'] = df.groupby(['NAME', 'ZIP_CODE'])['TRANSACTION_AMT'].transform('count')
    df = df[ ~( (df['count_all'] == 1) | (df['TRANSACTION_DT'].dt.year != actYear) )] 
    df['count_cur'] = df.groupby(['NAME', 'ZIP_CODE'])['TRANSACTION_AMT'].transform('count')

    df = df[ df['count_all'] > df['count_cur'] ] 

    df.set_index('ID', inplace=True)
    df.reset_index( inplace=True ) 

    df["IND"] = df.index + 1
    df["YEAR"] = df['TRANSACTION_DT'].dt.year
    df['TOT_CONTR'] = df['TRANSACTION_AMT'].cumsum()
    df['PercentileContrb'] = 0

    
    
    df['OrdinalRank'] = (actPercentile * df.index)/100
    df.loc[df['OrdinalRank'] < 0.5 , 'OrdinalRank'] = df['OrdinalRank'] + 0.5
    df['OrdinalRank'] = df.OrdinalRank.round().astype(int)

    list_contrb = []
    for index, row in df.iterrows():
        list_contrb.append(row['TRANSACTION_AMT'])
        print(sorted(list_contrb))
        df.set_value(index, 'PercentileContrb', sorted(list_contrb)[ row['OrdinalRank']-1 ] )
        
    
    print(df.shape)
    print(df ) 
    
    df.to_csv( fout,
              columns=['CMTE_ID','ZIP_CODE','YEAR','PercentileContrb','TOT_CONTR','IND'],
              sep='|', header=False,
              index=False) 
    return df

if __name__ == "__main__":
#     fcont = sys.argv[1]
#     fperc = sys.argv[2]
#     fout  = sys.argv[3]
    
    fcont = '/Users/ekinezgi/Downloads/itcont_test.txt'
    fperc = '/Users/ekinezgi/Downloads/percentile_test.txt'
    fout = '/Users/ekinezgi/Downloads/repeat_donors.txt'

    %time df = run(fcont, fperc,fout)
#     print( df[-1:] )
```
